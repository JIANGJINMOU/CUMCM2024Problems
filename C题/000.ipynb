```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures

plt.rcParams['font.sans-serif'] = ['SimHei']   # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False   # 用来正常显示负号

data = pd.read_csv('data.csv', encoding='GBK')

numeric_data = data[["23年产量", "销售单价中值", "种植成本"]]

# 销售量和销售价格的回归分析及散点图
X = numeric_data["23年产量"]
Y = numeric_data["销售单价中值"]
X = sm.add_constant(X)
# 使用多项式回归
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = sm.OLS(Y, X_poly).fit()
print(f"销售量和销售价格的回归结果：{model.summary()}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x="23年产量", y="销售单价中值", data=numeric_data)
plt.plot(X, model.predict(X_poly), color='red')
plt.xlabel("production volume")
plt.ylabel("sales price")
plt.show()

# 销售量和种植成本的回归分析及散点图
X = numeric_data["23年产量"]
Y = numeric_data["种植成本"]
X = sm.add_constant(X)
# 使用多项式回归
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = sm.OLS(Y, X_poly).fit()
print(f"销售量和种植成本的回归结果：{model.summary()}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x="23年产量", y="种植成本", data=numeric_data)
plt.plot(X, model.predict(X_poly), color='red')
plt.xlabel("production volume")
plt.ylabel("planting cost")
plt.show()

# 销售价格和种植成本的回归分析及散点图
X = numeric_data["销售单价中值"]
Y = numeric_data["种植成本"]
X = sm.add_constant(X)
# 使用多项式回归
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = sm.OLS(Y, X_poly).fit()
print(f"销售价格和种植成本的回归结果：{model.summary()}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x="销售单价中值", y="种植成本", data=numeric_data)
plt.plot(X, model.predict(X_poly), color='red')
plt.xlabel("sales price")
plt.ylabel("planting cost")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import Ridge, Lasso

plt.rcParams['font.sans-serif'] = ['SimHei']   # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False   # 用来正常显示负号

data = pd.read_csv('data.csv', encoding='GBK')

numeric_data = data[["23年产量", "销售单价中值", "种植成本"]]

# 销售量和销售价格的岭回归分析及散点图
X = numeric_data["23年产量"]
Y = numeric_data["销售单价中值"]
X = sm.add_constant(X)
ridge_model = Ridge(alpha=1.0)  # 调整 alpha 值来控制正则化强度
ridge_model.fit(X, Y)
print(f"销售量和销售价格的岭回归结果：{ridge_model.score(X, Y)}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x="23年产量", y="销售单价中值", data=numeric_data)
plt.plot(X, ridge_model.predict(X), color='red')
plt.xlabel("production volume")
plt.ylabel("sales price")
plt.show()

# 销售量和种植成本的lasso回归分析及散点图
X = numeric_data["23年产量"]
Y = numeric_data["种植成本"]
X = sm.add_constant(X)
lasso_model = Lasso(alpha=0.1)  # 调整 alpha 值来控制正则化强度
lasso_model.fit(X, Y)
print(f"销售量和种植成本的lasso回归结果：{lasso_model.score(X, Y)}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x="23年产量", y="种植成本", data=numeric_data)
plt.plot(X, lasso_model.predict(X), color='red')
plt.xlabel("production volume")
plt.ylabel("planting cost")
plt.show()

# 销售价格和种植成本的岭回归分析及散点图
X = numeric_data["销售单价中值"]
Y = numeric_data["种植成本"]
X = sm.add_constant(X)
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X, Y)
print(f"销售价格和种植成本的岭回归结果：{ridge_model.score(X, Y)}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x="销售单价中值", y="种植成本", data=numeric_data)
plt.plot(X, ridge_model.predict(X), color='red')
plt.xlabel("sales price")
plt.ylabel("planting cost")
plt.show()
```
